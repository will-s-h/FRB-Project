```python
import scipy.optimize as opt
```

```python
def logphi(logL, logP, logL0, d1, d2, n):
    return logP - d1*(logL-logP) - (1/n)*np.log10(1 + np.power(np.power(10, logL-logL0), n*(d2-d1)))
```

```python
xdata = Lprime2
ydata = phi2

logx = np.log10(xdata)
logy = np.log10(ydata)

guess2 = [1.4, 34.0, 0.16, 1.15, 1.0]
bounds = [[0.0, 30.0, 0.0, 0.0, 0.0], [10.0, 40.0, 10.0, 10.0, 10.0]]
optphi, pcov = opt.curve_fit(logphi, logx, logy, guess2, bounds=bounds)
```

```python
fig, ax = plt.subplots()
ax.scatter(np.log10(Lprime2), np.log10(phi2), label='Phi(L\') (k=2)')

modelLs = np.linspace(32.5, 35.5, 2000)
ax.plot(modelLs, logphi(modelLs, *optphi))

plt.legend()
plt.show()
```
